```python
n = 100 + 10 * 2
RGG = nx.random_geometric_graph(n, R)

pos = nx.get_node_attributes(RGG, "pos")

# find node near center (0.5,0.5)
dmin = 1
ncenter = 0
for n in pos:
    x, y = pos[n]
    d = (x - 0.5) ** 2 + (y - 0.5) ** 2
    if d < dmin:
        ncenter = n
        dmin = d

# color by path length from node near center
p = nx.single_source_shortest_path_length(RGG, ncenter)

plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(RGG, pos, alpha=0.4)
nx.draw_networkx_nodes(
    RGG,
    pos,
    nodelist=list(p.keys()),
    node_size=80,
    node_color=list(p.values()),
    cmap=plt.cm.Reds_r,
)

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
plt.axis("off")
plt.show()
```

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.
#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.
#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης.
#### Να σχολιαστούν τα αποτελέσματα.

```python
def cc(G):
  cc=nx.average_clustering(G)
  return cc
```

```python
# B1.1 Calculate and visualize the distribution of clustering coefficient for each node
def plot_clustering_coefficient_distribution(G, title):
    clustering_coefficients = list(nx.clustering(G).values())
    plt.figure()
    plt.hist(clustering_coefficients, bins=20)
    plt.title(f'Distribution of Clustering Coefficient for {title}')
    plt.xlabel('Clustering Coefficient')
    plt.ylabel('Frequency')
    plt.show()

plot_clustering_coefficient_distribution(REG, 'REG')
plot_clustering_coefficient_distribution(RGER, 'RG (ER)')
plot_clustering_coefficient_distribution(RGG, 'RGG')
plot_clustering_coefficient_distribution(SF, 'SF (BA)')
plot_clustering_coefficient_distribution(SW, 'SW (WS)')

```

```python
# B1.2 Calculate the mean and variance of the clustering coefficient
def calculate_mean_and_variance_clustering_coefficient(G):
    clustering_coefficients = list(nx.clustering(G).values())
    mean_cc = np.mean(clustering_coefficients)
    var_cc = np.var(clustering_coefficients)
    return mean_cc, var_cc

mean_reg_cc, var_reg_cc = calculate_mean_and_variance_clustering_coefficient(REG)
print(f'REG: Mean Clustering Coefficient = {mean_reg_cc:.4f}, Variance = {var_reg_cc:.4f}')

mean_rger_cc, var_rger_cc = calculate_mean_and_variance_clustering_coefficient(RGER)
print(f'RG (ER): Mean Clustering Coefficient = {mean_rger_cc:.4f}, Variance = {var_rger_cc:.4f}')

mean_rgg_cc, var_rgg_cc = calculate_mean_and_variance_clustering_coefficient(RGG)
print(f'RGG: Mean Clustering Coefficient = {mean_rgg_cc:.4f}, Variance = {var_rgg_cc:.4f}')

mean_sf_cc, var_sf_cc = calculate_mean_and_variance_clustering_coefficient(SF)
print(f'SF (BA): Mean Clustering Coefficient = {mean_sf_cc:.4f}, Variance = {var_sf_cc:.4f}')

mean_sw_cc, var_sw_cc = calculate_mean_and_variance_clustering_coefficient(SW)
print(f'SW (WS): Mean Clustering Coefficient = {mean_sw_cc:.4f}, Variance = {var_sw_cc:.4f}')

```

```python
# B.1.3 Find the nodes with the smallest and largest clustering coefficient
def find_min_max_clustering_coefficient_nodes(G):
    clustering_coefficients = nx.clustering(G)
    min_cc_node = min(clustering_coefficients, key=clustering_coefficients.get)
    max_cc_node = max(clustering_coefficients, key=clustering_coefficients.get)
    return min_cc_node, clustering_coefficients[min_cc_node], max_cc_node, clustering_coefficients[max_cc_node]

min_reg_node, min_reg_cc_value, max_reg_node, max_reg_cc_value = find_min_max_clustering_coefficient_nodes(REG)
print(f'REG: Node with min CC = {min_reg_node} ({min_reg_cc_value:.4f}), Node with max CC = {max_reg_node} ({max_reg_cc_value:.4f})')

min_rger_node, min_rger_cc_value, max_rger_node, max_rger_cc_value = find_min_max_clustering_coefficient_nodes(RGER)
print(f'RG (ER): Node with min CC = {min_rger_node} ({min_rger_cc_value:.4f}), Node with max CC = {max_rger_node} ({max_rger_cc_value:.4f})')

min_rgg_node, min_rgg_cc_value, max_rgg_node, max_rgg_cc_value = find_min_max_clustering_coefficient_nodes(RGG)
print(f'RGG: Node with min CC = {min_rgg_node} ({min_rgg_cc_value:.4f}), Node with max CC = {max_rgg_node} ({max_rgg_cc_value:.4f})')

min_sf_node, min_sf_cc_value, max_sf_node, max_sf_cc_value = find_min_max_clustering_coefficient_nodes(SF)
print(f'SF (BA): Node with min CC = {min_sf_node} ({min_sf_cc_value:.4f}), Node with max CC = {max_sf_node} ({max_sf_cc_value:.4f})')

min_sw_node, min_sw_cc_value, max_sw_node, max_sw_cc_value = find_min_max_clustering_coefficient_nodes(SW)
print(f'SW (WS): Node with min CC = {min_sw_node} ({min_sw_cc_value:.4f}), Node with max CC = {max_sw_node} ({max_sw_cc_value:.4f})')
```

### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.
#### B2.2 Να υπολογιστεί η διάμετρος (συνάρτηση diameter) και η ακτίνα (συνάρτηση radius).
#### Για τα παραπάνω πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως παρουσιάζονται στη θεωρία.

B2.1


```python
def compute_path_lengths(G):
    path_lengths = dict(nx.all_pairs_shortest_path_length(G))
    return path_lengths
```

```python
def avg(G):
    path_lengths = compute_path_lengths(G)
    sum = 0
    for path in path_lengths:
      sum += path
    avg = sum/len(path_lengths)
    return avg
```

```python
def var(G):
  path_lengths = compute_path_lengths(G)
  mu = avg(G)
  sum = 0
  for path in path_lengths:
    sum += (path - mu)**2
  var = sum/len(path_lengths)
  return var
```

```python
for gr_name, gr in graph_dict.items():
    # Compute shortest path lengths for all pairs of nodes
    path_lengths_dict = dict(nx.all_pairs_shortest_path_length(gr))

    # Extract all path lengths into a list, excluding paths from a node to itself
    path_lengths_list = [length for source, targets in path_lengths_dict.items() for target, length in targets.items() if source != target]

    if path_lengths_list:
        mean = np.mean(path_lengths_list)
        var = np.var(path_lengths_list)
        print(f'{gr_name}: Mean is: {mean:.4f}. Variance is: {var:.4f}.')
    else:
        print(f'{gr_name}: No paths found or graph is disconnected.')
```

#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python
def find_nodes(G):
  try:
          diameter = nx.diameter(G)
          print(f"\nAnalyzing {name} for nodes at diameter distance...")

          # Compute all-pairs shortest paths
          shortest_paths = dict(nx.shortest_path_length(G))

          max_count = -1
          node_with_max_count = None
          nodes_at_diameter = []

          for source_node, paths in shortest_paths.items():
              count_at_diameter = sum(1 for target_node, length in paths.items() if length == diameter)
              if count_at_diameter > max_count:
                  max_count = count_at_diameter
                  node_with_max_count = source_node
                  nodes_at_diameter = [target_node for target_node, length in paths.items() if length == diameter]

          if node_with_max_count is not None:
              print(f"  Node with most nodes at diameter distance ({diameter}): {node_with_max_count}")
              print(f"  Number of nodes at diameter distance: {max_count}")
              print(f"  Nodes at diameter distance from {node_with_max_count}: {nodes_at_diameter}")
          else:
              print(f"  No nodes found at diameter distance.")
  except nx.NetworkXNoPath:
      print(f"  Graph is not connected. Cannot calculate diameter and analyze nodes at diameter distance.")
  except nx.NetworkXError as e:
      print(f"  Error analyzing {name} - {e}")
```

```python
for name, graph in graph_dict.items():
   find_nodes(graph)
```

## Γ) Mετρικές κεντρικότητας κόμβων

### Γ1. Κεντρικότητα βαθμού (Degree centrality)
##### Γ1.1 Να αναπαρασταθεί η κατατανομή βαθμών κόμβων για κάθε τοπολογία.
##### Γ1.2 Να υπολογιστεί ο μέσος βαθμός κόμβου καθώς και η διασπορά των βαθμών κόμβου για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_degree(G):
    degrees = [G.degree(n) for n in G.nodes()]
    mean_degree = np.mean(degrees)
    var_degree = np.var(degrees)
    return degrees, mean_degree, var_degree
```

```python
# Γ1.1 & Γ1.2 Calculate and visualize degree distribution, mean and variance
for name, graph in graph_dict.items():
    degrees, mean_degree, var_degree = find_degree(graph)

    print(f"\n{name}:")
    print(f"  Mean Degree: {mean_degree:.4f}")
    print(f"  Variance of Degree: {var_degree:.4f}")

    plt.figure()
    plt.hist(degrees, bins=20)
    plt.title(f'Degree Distribution for {name}')
    plt.xlabel('Degree')
    plt.ylabel('Frequency')
    plt.show()
```

### Γ4. Kεντρικότητα Katz
#### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία.
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx.
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.
#### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.
#### Να σχολιαστούν τα αποτελέσματα.

```python
def find_katz(G):
    # Calculate the largest eigenvalue of the adjacency matrix
    try:
        eigenvalues = nx.adjacency_spectrum(G)
        lambda_max = max(abs(eigenvalues))
        alpha = 1/lambda_max - 0.01

        # Calculate Katz centrality
        katz_centralities = list(nx.katz_centrality(G, alpha=alpha).values())

        print(f"\n{name}:")
        mean_katz = np.mean(katz_centralities)
        var_katz = np.var(katz_centralities)
        print(f"  Mean Katz Centrality: {mean_katz:.4f}")
        print(f"  Variance of Katz Centrality: {var_katz:.4f}")

        plt.figure()
        plt.hist(katz_centralities, bins=20)
        plt.title(f'Katz Centrality Distribution for {name}')
        plt.xlabel('Katz Centrality')
        plt.ylabel('Frequency')
        plt.show()
    except nx.NetworkXError as e:
        print(f"\n{name}: Error calculating Katz centrality - {e}")
```

```python
# Γ4.1 & Γ4.2 Calculate and visualize Katz centrality distribution, mean and variance
for name, graph in graph_dict.items():
  find_katz(graph)
```

#### Να προσδιοριστούν, για κάθε τοπολογία, οι πιο κεντρικοί κόμβοι ως προς το βαθμό, την εγγύτητα, την ενδιαμεσική κεντρικότητα και την κεντρικότητα Katz. Να σχολιαστούν τα αποτελέσματα.

```python
def find_central(G, name):
    # Degree Centrality
    degree_centrality = nx.degree_centrality(G)
    max_degree_node = max(degree_centrality, key=degree_centrality.get)
    print(f"\nMost central node in {name} by Degree Centrality: Node {max_degree_node} ({degree_centrality[max_degree_node]:.4f})")

    # Closeness Centrality
    try:
        closeness_centrality = nx.closeness_centrality(G)
        max_closeness_node = max(closeness_centrality, key=closeness_centrality.get)
        print(f"Most central node in {name} by Closeness Centrality: Node {max_closeness_node} ({closeness_centrality[max_closeness_node]:.4f})")
    except nx.NetworkXError as e:
        print(f"\n{name}: Cannot calculate Closeness Centrality - {e}")

    # Betweenness Centrality
    betweenness_centrality = nx.betweenness_centrality(G)
    max_betweenness_node = max(betweenness_centrality, key=betweenness_centrality.get)
    print(f"Most central node in {name} by Betweenness Centrality: Node {max_betweenness_node} ({betweenness_centrality[max_betweenness_node]:.4f})")

    # Katz Centrality (need to calculate alpha first)
    try:
        eigenvalues = nx.adjacency_spectrum(G)
        lambda_max = max(abs(eigenvalues))
        alpha = 1/lambda_max - 0.01
        katz_centrality = nx.katz_centrality(G, alpha=alpha)
        max_katz_node = max(katz_centrality, key=katz_centrality.get)
        print(f"Most central node in {name} by Katz Centrality: Node {max_katz_node} ({katz_centrality[max_katz_node]:.4f})")
    except nx.NetworkXError as e:
        print(f"\n{name}: Cannot calculate Katz Centrality - {e}")

# Call the find_central function for each graph
for name, graph in graph_dict.items():
    find_central(graph, name)
```

```python
n = 120 # network size
d = 4   # initial lattice degree

gp_values = np.arange(0, 1.1, 0.1) # g_p from 0 to 1 with step 0.1

avg_path_lengths = []
avg_clustering_coefficients = []

for gp in gp_values:
    # Create Watts-Strogatz graph
    # The seedno should be used here for reproducibility if it was used in previous related graph generations
    # nx.watts_strogatz_graph does not have a seed parameter directly in the function call,
    # but random.seed() is called at the beginning of the notebook so it applies globally.
    SW_graph = nx.watts_strogatz_graph(n, d, gp)

    # Calculate average clustering coefficient
    avg_cc = nx.average_clustering(SW_graph)
    avg_clustering_coefficients.append(avg_cc)

    # Calculate average path length
    if nx.is_connected(SW_graph):
        avg_spl = nx.average_shortest_path_length(SW_graph)
        avg_path_lengths.append(avg_spl)
    else:
        avg_path_lengths.append(np.nan) # Assign NaN if not connected

# Plot Average Path Length vs g_p
plt.figure(figsize=(10, 6))
plt.plot(gp_values, avg_path_lengths, marker='o')
plt.title('Average Path Length vs. Rewiring Probability (g_p)')
plt.xlabel('Rewiring Probability (g_p)')
plt.ylabel('Average Path Length')
plt.grid(True)
plt.show()

# Plot Average Clustering Coefficient vs g_p
plt.figure(figsize=(10, 6))
plt.plot(gp_values, avg_clustering_coefficients, marker='o', color='orange')
plt.title('Average Clustering Coefficient vs. Rewiring Probability (g_p)')
plt.xlabel('Rewiring Probability (g_p)')
plt.ylabel('Average Clustering Coefficient')
plt.grid(True)
plt.show()
```

The network behavior changes significantly with the rewiring probability g_p in the Watts-Strogatz model:

    For g_p = 1: The network transforms into a random graph, exhibiting a very low average clustering coefficient (around 0.03) and a very low average path length (around 3.59). This indicates a highly randomized structure where local connections are sparse, but any two nodes are relatively close.
    For intermediate values (e.g., 0.1 to 0.5): This range demonstrates the "small-world" phenomenon. Even with small values of g_p, the average path length drops dramatically (e.g., to approximately 5.02 at g_p = 0.1) due to the introduction of a few long-range shortcuts. Crucially, the average clustering coefficient remains significantly higher than that of a purely random graph, showing that local connectivity is largely preserved.
